```python
#Elimino culumna Online_Support_Usage, todos sus datos son NaN
df=df.drop(columns=['Online_Support_Usage'])
df
```

```python
#Eliminar duplicados
df=df.drop_duplicates()
df
```

Renombrar nombres de columnas. 

```python
df2 = df.rename(columns= {'User_ID': 'Usuario', 'Age': 'Edad', 'Gender': 'GÃ©nero', 'Technology_Usage_Hours': 'Horas_de_uso _de_la_tecnologÃ­a', 'Social_Media_Usage_Hours': 'Horas_de_uso_de_las_redes_sociales', 'Gaming_Hours': 'Horas_de_juego', 'Screen_Time_Hours': 'Horas_de_tiempo_en_pantalla', 'Mental_Health_Status': 'Estado_de_salud_mental', 'Stress_Level': 'Nivel_de_estrÃ©s', 'Sleep_Hours': 'Horas_de_sueÃ±o', 'Physical_Activity_Hours': 'Horas_de_actividad_fÃ­sica', 'Support_Systems_Access': 'Acceso_a_sistemas_de_soporte', 'Work_Environment_Impact': 'Impacto_en_el_entorno_laboral'})
df2
```

Renombrar datos de columnas. 

```python
#GÃ©nero
gen={
    'Female': 'Mujer',
    'Male': 'Hombre',
    'Other': 'Otro'
}
#Salud mental
sm={
    'Excellent': 'Excelente',
    'Good': 'Buena',
    'Fair': 'Neutral',
    'Poor': 'Mala'
}
#Nivel de estrÃ©s
ne={
    'Low': 'Bajo',
    'Medium': 'Medio',
    'High': 'Alto'
}
#Acceso a sistemas de soporte
ss={
    'Yes': 'Si'
}
#Impacto en el entorno laboral
el={
    'Negative': 'Negativo',
    'Positive': 'Positivo'
}
```

```python
df2['GÃ©nero']=df2['GÃ©nero'].replace(gen)
df2['Estado_de_salud_mental']=df2['Estado_de_salud_mental'].replace(sm)
df2['Nivel_de_estrÃ©s']=df2['Nivel_de_estrÃ©s'].replace(ne)
df2['Acceso_a_sistemas_de_soporte']=df2['Acceso_a_sistemas_de_soporte'].replace(ss)
df2['Impacto_en_el_entorno_laboral']=df2['Impacto_en_el_entorno_laboral'].replace(el)
df2
```

Eliminamos los invalid.

```python
#Contamos los invalid 
lista_col=df2.columns
for i in lista_col:
    print(f"En la columna {i} los invalid son: {df2[df2[i] == 'invalid'].shape[0]}")
```

```python
#Eliminamos los invalid
for i in lista_col:
    df2=df2[df2[i] != 'invalid']
df2
```

```python
lista_col=df2.columns
for i in lista_col:
    print(f"En la columna {i} los invalid son: {df2[df2[i] == 'invalid'].shape[0]}")
```

Remplazamos los espacios vacÃ­os. 

```python
#Cambiamos de objeto a float los datos numÃ©ricos, el cÃ³digo se salta los espacios vacÃ­os 
df2['Edad'] = pd.to_numeric(df2['Edad'], errors='coerce')
df2['Horas_de_uso _de_la_tecnologÃ­a'] = pd.to_numeric(df2['Horas_de_uso _de_la_tecnologÃ­a'], errors='coerce')
df2['Horas_de_uso_de_las_redes_sociales'] = pd.to_numeric(df2['Horas_de_uso_de_las_redes_sociales'], errors='coerce')
df2['Horas_de_sueÃ±o'] = pd.to_numeric(df2['Horas_de_sueÃ±o'], errors='coerce')
df2['Horas_de_actividad_fÃ­sica'] = pd.to_numeric(df2['Horas_de_actividad_fÃ­sica'], errors='coerce')
df2
```

```python
df2.isnull().sum()
```

```python
df2.info()
```

```python
#Remplazamos los valores numÃ©ricos por su promedio.
df2['Edad'].fillna(df2['Edad'].mean(), inplace=True)
df2['Horas_de_uso _de_la_tecnologÃ­a'].fillna(df2['Horas_de_uso _de_la_tecnologÃ­a'].mean(), inplace=True)
df2['Horas_de_uso_de_las_redes_sociales'].fillna(df2['Horas_de_uso_de_las_redes_sociales'].mean(), inplace=True)
df2['Horas_de_sueÃ±o'].fillna(df2['Horas_de_sueÃ±o'].mean(), inplace=True)
df2['Horas_de_actividad_fÃ­sica'].fillna(df2['Horas_de_actividad_fÃ­sica'].mean(), inplace=True)
df2['Horas_de_juego'].fillna(df2['Horas_de_juego'].mean(), inplace=True)
df2['Horas_de_tiempo_en_pantalla'].fillna(df2['Horas_de_tiempo_en_pantalla'].mean(), inplace=True)
df2
```

```python
#Elimino esta columna  Usuario porque no aporta nada a la investigaciÃ³n. 
df3=df2.drop(columns=['Usuario'])
df3
```

```python
#Remplazamos espacios vacÃ­os de las columnas tipo objeto, por datos "Neutrales".
df3['GÃ©nero'].fillna("Otro", inplace=True)
df3['Estado_de_salud_mental'].fillna("Neutral", inplace=True)
df3['Nivel_de_estrÃ©s'].fillna("Medio", inplace=True)
df3['Impacto_en_el_entorno_laboral'].fillna("Neutral", inplace=True)
```

```python
df3.isnull().sum()
```

```python
#Elimino los NaN restantes, son de la columna 'Acceso_a_sistemas_de_soporte' y no pueden ser rellenados porque solo hay datos de sÃ­ y no. 
df4=df3.dropna()
```

```python
df4.isnull().sum()
```

```python
df4 = df4.reset_index(drop=True)
```

```python
#Cambiamos a entero
df4['Edad']=df4['Edad'].astype(int)
```
